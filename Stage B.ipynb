### Regression In Machine Learning

```python
# importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# importing the dataset
df = pd.read_csv('energydata_complete.csv')
```

```python
#first 5 rows
df.head()
```

### A Simple linear model between T2 and T6

```python
x = df[['T2']]
y = df['T6']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x, y)
```

```python
#evaluation using R^2 
r2 = linear_model.score(x, y)
print('The R^2 is',round(r2,2))
```

### Multiple Linear Regression

```python
#dropping some features
df = df.drop(['date', 'lights'], axis = 1)
```

```python
df.head()
```

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
norm_df = pd.DataFrame(minmax_scaler.fit_transform(df), columns=df.columns)

#defining the train and target features
X= norm_df.drop(columns = ['Appliances'])
y = norm_df['Appliances']
```

```python
#splitting into test and train for validation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
x_test
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
pred = linear_model.predict(x_test)
```

#### Evaluation

##### MAE(Mean Absolute Error)

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, pred)
round(mae, 2)
```

##### RSS(Residual Sum of Squares)

```python
rss = np.sum(np.square(y_test - pred))
round(rss, 2)
```

##### RMSE(Root Mean Squared Error)

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, pred))
round(rmse, 3)
```

##### Coefficient of Determination

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred)
round(r2_score, 2)
```

##### Linear model Features weights

```python
#comparing the effects of regularisation 
def get_weights_df(model, feat, col_name): 
    #this function returns the weight of every feature 
    weights = pd.Series(model.coef_, feat.columns).sort_values() 
    weights_df = pd.DataFrame(weights).reset_index() 
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3) 
    
    return weights_df 

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```

```python
linear_model_weights
```

### Ridge Regression Model

```python
from sklearn.linear_model import Ridge 
ridge_model = Ridge(alpha=0.4)
ridge_model.fit(x_train, y_train)
```

```python
#obtain predictions
ridge_pred = ridge_model.predict(x_test)
```

#### Evaluation

##### RMSE(Root Mean Squared Error)

```python
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))

round(ridge_rmse, 3)
```

### Lasso Regression Model

```python
from sklearn.linear_model import Lasso 
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

#obtain predictions
lasso_pred = lasso_reg.predict(x_test)
```

##### Lasso regression model Features weights

```python
#getting the weights
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

#### Evaluation

##### RMSE(Root Mean Squared Error)

```python
# Root Mean Square Error (RMSE) 
lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso_pred))

round(lasso_rmse, 3)
```
