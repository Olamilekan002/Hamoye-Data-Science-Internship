```python
#importing necessary libraries
import numpy as np
import pandas as pd
```

```python
#downloading and importing dataset
# url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'
# df = pd.read_csv(url, encoding ='latin-1')

df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin-1')
```

```python
#A brief view of the dataset
df.head()
```

#### Question 11: The total sum of the Animal Fat produced in 2014 and 2017 respectively

```python
df.groupby('Item').sum().loc[['Animal fats']][['Y2014', 'Y2017']]
```

#### Question 12: The mean and the std across the whole dataset for the whole dataset for the year 2015

```python
df.describe()
```

```python
#results rounded to 3 decimal place
df.describe().loc[['mean', 'std']][['Y2015']].round(3)
```

#### Question 13: The total number and percentage of missing data in 2016

```python
missing = df.isnull().sum()['Y2016']
percent = 100 * (missing / len(df))
print('The total number of missing data for 2016 is {} which is {:.2f}%'.format(missing, percent))
```

#### Question 14: The year with the highest correlation with 'Element Code'

```python
df.corr()
```

```python
years = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
df.corr().loc[['Element Code']][years]\
        .sort_values(by = 'Element Code', axis = 1, ascending = False)
```

```python
#Year 2014 has the highest correlation
```

#### Question 15: The year with the highest sum of Import Quantity

```python
df.groupby('Element').sum()
```

```python
df.groupby('Element').sum().loc[['Import Quantity']][years]\
                    .sort_values(by = 'Import Quantity', axis = 1, ascending = False)
```

```python
#Year 2017 has the highest sum of import quantity
```

#### Question 16:  The total number of the sum of production in 2014

```python
df.groupby('Element').sum().loc[['Production']][['Y2014']]
```

#### Question 17: The Element with highest sum in 2018

```python
df[['Y2018', 'Element']].groupby('Element').sum()\
        .sort_values(by='Y2018', ascending = False).iloc[[0]]
```

```python
# Domestic supply quantity has the highest
```

#### Question 18: The 3rd lowest sum in 2018

```python
df[['Y2018', 'Element']].groupby('Element').sum().sort_values(by='Y2018').iloc[[2]]
```

```python
# Protein supply quantity (g/capita/day) is the 3rd lowest in 2018
```

#### Question 19: The total Import Quantity in Algeria in 2018

```python
df.groupby(['Area', 'Element']).sum().loc[[('Algeria', 'Import Quantity')]][['Y2018']]
```
